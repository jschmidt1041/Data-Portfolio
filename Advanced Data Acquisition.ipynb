## C1. Dashboard Purpose and Functions
Heart disease is the leading cause of death in the United States (source: CDC). The purpose of the dashboard, "Assessing the Health of Our Patients", is to analyze trends and similarities between our patients and patients with known heart disease so that we may better serve our patients in terms of both preventive and active treatment measures. In other words, by analyzing traits of patients with heart disease, we can better detect and prepare for treating heart disease in our patients.

Essentially, the variables in the additional dataset from the CDC offers insights into patients with heart disease. We can utilize this data to compare it with our patients' data and get more insights like first, assess the overall health of our patient population compared with the health of heart disease patients, and second, look for trends and characteristics of heart disease patients to determine what to look for when preventing and detecting heart disease in our own patients. For example, we know that heart disease patients tend to have higher BMIs, so we can more closely monitor our patients with above average BMIs for heart disease.

## C5. Results of the Analysis
The results of the analysis are good such that we were able to draw insights from the dashboard. Since we were looking for risk factors for heart disease in part C1, we can see the results of the dashboard show exactly that. We can see that our patient population is similar to the CDC patient population with heart disease in terms of BMI, but we have fewer patients with an overweight BMI - this is good, but it is also something that should be addressed.

Secondly, we can see that a larger percentage of heart disease patients have high blood pressure than our patients, which is a good sign. However, it is also important to know because we are able to keep a closer eye on our patients who have high blood pressure since they are at a higher risk of heart disease.

Third, we can see that more patients with heart disease have diabetes than our patient population, which is similar in effect to the high blood pressure noted above: it is a good sign, but we also need to keep a close eye on those who do have diabetes.

Finally, we get a generalized look at some of the demographics of our patient population which can be useful. Namely, we have age, gender, and martial status. The most common age of our patients is 47, we have more female patients than male, and our patients' marital statuses are almost even across the board.

## C6. Limitations of the Data Analysis
Some limitations of my analysis are the number of records in both datasets - the analysis would be more robust with more data. There also were some variables missing from the WGU Medical Dataset that would have been useful in comparison to the Kaggle dataset in detecting and analyzing heart disease, specifically Smoking and even heart disease/heart disease risk itself. This means that we had to analyze associated risk factors of heart disease/heart disease risk instead of being able to analyze it directly, thereby limiting the effectiveness of the analysis.
